```python
# Percentage of missing values of all the columns
def get_percentage_missing(application):
    """ Calculates percentage of NaN values in DataFrame
    :param application: Pandas DataFrame object
    :return: float
    """
    num = application.isnull().sum()
    den = len(application)
    return round(num/den, 2)
```

## Data Cleaning

```python
application.isnull()
```

```python
# Check the missing values in the table 
def missing_values_table(application): 
    mis_val = application.isnull().sum()
    mis_val_percent = 100 * application.isnull().sum()/len(application)
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'})
    return mis_val_table_ren_columns
```

```python
# Removing the columns with high missing percentage (>50%)
application = application.drop(application.columns[application.apply(lambda col: col.isnull().sum()/len(application) > 0.50)], axis=1)
```
